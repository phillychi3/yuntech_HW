```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from sympy import *
```

```python
%matplotlib inline

def f(x):
    return x**2-np.log(x)

x=np.linspace(-1000,1000,201)
y=f(x)
plt.plot(x, y)
```

```python
def f(x):
    return 2*x - 1/x

x=np.linspace(-1000,1000,201)
y=f(x)
plt.plot(x, y)
```
