```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# laod data
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
df.describe()
```

```python
# total Platform games
df['Platform'].value_counts()
```

```python
plt.bar(df['Platform'].value_counts().index[:10], df['Platform'].value_counts().values[:10])
```

```python
platG = pd.crosstab(df.Platform,df.Genre)
platall = platG.sum(axis=1).sort_values(ascending = False)
plt.figure(figsize=(8,7))
sns.barplot(y = platall.index, x = platall.values, orient='h')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(10,10))
sns.heatmap(platG, cmap='Blues', annot=True, fmt='d')
plt.show()
```
